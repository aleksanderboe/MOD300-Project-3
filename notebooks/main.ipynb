```python
def create_random_point(x_lower, y_lower, z_lower, x_upper, y_upper, z_upper) -> tuple:
    """
    Creates an random point within the 3D simulation box.

    :return: A tuple (x, y, z) representing the coordinates of the 
             randomly generated point inside the simulation box. 
    """
    x = np.random.uniform(x_lower, x_upper)
    y = np.random.uniform(y_lower, y_upper)
    z = np.random.uniform(z_lower, z_upper)
    return (x,y,z)

create_random_point(x_0, y_0, z_0, x_1, y_1, z_1)
```
